```python
from pytorch3d.vis import plot_scene
from pytorch3d.renderer.cameras import PerspectiveCameras
from pytorch3d.transforms import quaternion_to_matrix
import pandas as pd
import torch
```

```python
# load the csv
df = pd.read_csv('output/C0044.csv')
df = df.iloc[::1, -4:]
focal_length = []
principal_points = []
R = []
T = []

def quat2rotm(q_vec: list):
    q = torch.Tensor(q_vec) # w, x, y, z # w is real
    m = quaternion_to_matrix(q)
    return m

for k in range(len(df)):
    focal_length.append((1., 1.))
    principal_points.append((0., 0.))    
    R.append(quat2rotm(df.iloc[k, :4].values))
    T.append(torch.Tensor([0., 0., k]))

focal_length = torch.Tensor(focal_length)
principal_points = torch.Tensor(principal_points)
R = torch.stack(R)
T = torch.stack(T)

cams = PerspectiveCameras(
    focal_length=focal_length,
    principal_point=principal_points,
    R=R,
    T=T
)

plot_scene({"COLMAP Reconstruction (using SiLK keypoints)": {"cameras": cams}}, cameras=True)
```
